Variance analysis of Actual and Target

```python
data_comb=data_Actuals.merge(data_Target,on=['Plant'])
```

```python
data_comb.head()
```

```python
data_comb.dropna()
```

```python
data_comb=data_comb[data_comb['Quantity']>=0]
```

```python
data_comb=data_comb[data_comb['Material Description']==data_comb['Material Number']]
```

```python
data_comb.describe()
```

```python
data_comb.head()
```

```python
data_comb=data_comb.drop(data_comb[['Material Number']],axis=1)
```

```python
data=data_comb[['Material Description','Plant']]
```

```python
data_comb=data_comb.merge(data_Price,left_on='Material Description',right_on='Material Description')
```

```python
data_comb.head()
```

```python
data_comb['Total price Actual']=data_comb['Quantity']*data_comb['Amount in LC']
```

```python
data_comb['Total price Target']=data_comb['Target Quantity']*data_comb['Target Value in LC']
```

```python
data_comb.head()
```

```python
data_comb=data_comb.drop(data_comb[['Target Value in LC']],axis=1)
```

```python
data_comb=data_comb.drop(data_comb[['Target Quantity']],axis=1)
data_comb=data_comb.drop(data_comb[['Amount in LC']],axis=1)
data_comb=data_comb.drop(data_comb[['Quantity']],axis=1)
```

```python
ax=data_comb.plot(x="Period_x",y='Total price Actual',legend=False)
ax2 = ax.twinx()
data_comb.plot(x="Period_x", y="Total price Target", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```
